# Importing all libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
```

```python
from sklearn.ensemble import StackingClassifier, RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

# Data cleaning

```python
df = pd.read_csv(r"D:\Study\Testing dataset\CSV data\blood_samples_dataset_test.csv")
```

```python
df.head(10)
```

```python
df.isnull()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.iloc[:, 0:-1]
Y = df.iloc[:, -1]
```

```python
print(X.shape)
print(Y.shape)
```

```python
Y = LabelEncoder().fit_transform(Y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, train_size=0.8, shuffle=True)
```

# Creating model

```python
base_model = ([('gb', GradientBoostingClassifier()), ('lr', LogisticRegression())])
```

```python
stack = StackingClassifier(estimators= base_model, final_estimator= RandomForestClassifier())
```

```python
pipe = Pipeline([('scaler', StandardScaler()), ('model', stack)])
```

```python
parameter = {
    'model__gb__loss': ['log_loss'],
    'model__gb__learning_rate': [0.1, 0.001, 0.0001],
    'model__lr__penalty': ['l2'],
    'model__lr__C': [2.0],
    'model__final_estimator__n_estimators': [50, 100, 150, 200],  # Corrected `n_estimators`
    'model__final_estimator__criterion': ["log_loss"],
    'model__final_estimator__max_depth': [20],
    'model__final_estimator__min_samples_leaf': [10],
    'model__final_estimator__min_samples_split': [4],
    'model__final_estimator__max_features': ["sqrt", "log2"],
}
```

```python
random_search_cv = RandomizedSearchCV(pipe, param_distributions= parameter, n_iter = 2, cv = 2, n_jobs = -1, verbose=2, error_score='raise')
```

```python
random_search_cv.fit(x_train, y_train)
```

```python
random_search_cv.score(x_test, y_test)
```

```python
random_search_cv.best_estimator_
```

```python
predict = random_search_cv.predict(x_test)
```

```python
cm = classification_report(y_test, predict)
print(cm)
```

```python
import pickle

# Save the model
with open("random_search_cv.pkl", "wb") as file:
    pickle.dump(random_search_cv, file)
```
